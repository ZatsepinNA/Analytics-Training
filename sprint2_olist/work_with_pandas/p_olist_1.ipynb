##Микроввыод: Все нулевые значения в дате доставки связаны с тем, что заказ просто не был доставлен. В данном случае, исключаем нулевые значения ничем их не заполняя, так как они не несут информации. Разница в количестве NaN значений в столбцах, тем или иным образом, связанных с доставкаой объясняется тем, что по многообразию статусов можно сказать о том, что некоторые из заказов отвалились на каком либо этапе доставки. Впрочем, нас интересует только доставленные заказы.

```python
# Как видно - новый датасет нормализован и готов к работе. 
cleaned2_orders = orders.copy().dropna()
date_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_columns:
    cleaned2_orders[col] = pd.to_datetime(cleaned2_orders[col])
print(cleaned2_orders.info())
```

```python
print(orders['order_purchase_timestamp'].info())
print('==================')
print(cleaned2_orders['order_purchase_timestamp'].info())
```
